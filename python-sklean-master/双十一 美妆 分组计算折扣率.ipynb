```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('双十一淘宝美妆数据.csv',
               encoding='gbk',parse_dates=['update_time'])
# parse_dates参数：将csv中的时间字符串转换成日期格式

df.sort_values(['id','update_time'],inplace=True)

df.reset_index(drop=True,inplace=True)

df
```

```python
def f(x):
    before_11 = x[x['update_time'] < '2016-11-11']
    on_11 = x[x['update_time'] == '2016-11-11']
    after_11 = x[x['update_time'] > '2016-11-11']
    
    x['双11前'] = not before_11.empty
    x['双11当天'] = not on_11.empty
    x['双11后'] = not after_11.empty
    
    return x

fl = df.groupby('id').apply(f)
fl
```

```python
fl.loc[(fl['双11前'] == True) &
        (fl['双11当天'] == True) &
        (fl['双11后'] == True),'类别'] = "A"

fl.loc[(fl['双11前'] == True) &
        (fl['双11当天'] == True) &
        (fl['双11后'] == False),'类别'] = "B"
        
fl.loc[(fl['双11前'] == False) &
        (fl['双11当天'] == True) &
        (fl['双11后'] == True),'类别'] = "C"

fl.loc[(fl['双11前'] == False) &
        (fl['双11当天'] == True) &
        (fl['双11后'] == False),'类别'] = "D"

fl.loc[(fl['双11前'] == True) &
        (fl['双11当天'] == False) &
        (fl['双11后'] == False),'类别'] = "E"

fl.loc[(fl['双11前'] == False) &
        (fl['双11当天'] == False) &
        (fl['双11后'] == True),'类别'] = "F"
        
fl.loc[(fl['双11前'] == True) &
        (fl['双11当天'] == False) &
        (fl['双11后'] == True),'类别'] = "F"
        
        
        
A = fl.loc[fl['类别'] == 'A']
A
```

```python
def dis(x):
    p11 = x.loc[x['update_time']=='2016-11-11','price'].mean()
    pmax = x.loc[x['update_time']!='2016-11-11','price'].max()
    x['折扣率'] = p11/pmax
    print(p11)
    return x[['id','title','店名','类别','折扣率']].iloc[0]

A_zkl=A.groupby('id',as_index=False).apply(dis)
A_zkl.sort_values('折扣率')
```
