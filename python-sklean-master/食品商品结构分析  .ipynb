```python
from sklearn.cluster import KMeans,SpectralClustering
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.metrics import silhouette_score as ss
```

```python
zhuanqu.corr()
```

```python
x = zhuanqu[['价格','总销量','评价数','预估交易占比','预估交易增幅']]
x
```

```python
for i in x.columns[:-1]:
    x[[i]] = MinMaxScaler().fit_transform(x[[i]])
```

```python
x
```

```python
x[['预估交易增幅']] = StandardScaler().fit_transform(x[['预估交易增幅']])
```

```python
x
```

```python
k_list = [i for i in range(2,15)]
```

```python
k_list
```

```python
score = pd.Series()
for i in k_list:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(x)
    labels = kmeans.labels_
    score[str(i)] = ss(x,labels,metric='euclidean')
score.plot()
```

```python
score[1:].max()
```

```python
kmeans = KMeans(n_clusters = 2)
kmeans.fit(x)
kmeans.cluster_centers_

x['label'] = kmeans.labels_
x_1 = zhuanqu[['价格','总销量','评价数','预估交易占比','预估交易增幅']]
x_1['label'] = kmeans.labels_
```

```python
x_1
```

```python
x_2 = x[x.label == 0]
score2 = pd.Series()
for i in k_list:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(x_2.iloc[:,:-1])
    labels = kmeans.labels_
    score2[str(i)] = ss(x_2.iloc[:,:-1],labels,metric='euclidean')
score2.plot()

#k=8时 轮廓系数最大，不到0.475。这个轮廓系数在可接受的范围内，但是K值仍然太大了。取5也可以其实。
#事实上，轮廓系数虽然越接近1越好，但实际中高于0.3或0.4就可以接受
#试试组内平方和
```

```python
def kmean_sse(x,clusters):
    k = range(1,clusters+1)
    total_sse = []
    for i in k:
        sse = []
        kmeans = KMeans(n_clusters = i)
        kmeans.fit(x)
        labels = kmeans.labels_
        centers = kmeans.cluster_centers_
        for label in set(labels):
            sse.append(np.sum((x.loc[labels==label,] - centers[label,:])**2))
        total_sse.append(np.sum(sse))
    
    plt.plot(k,total_sse,'b*-')
    plt.show()
```

```python
kmean_sse(x_2[['价格','总销量','评价数','预估交易占比','预估交易增幅']],15)
```

```python
kmean_sse(x_1[['价格','总销量','评价数','预估交易占比','预估交易增幅']],15)
```
